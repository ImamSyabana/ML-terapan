melakukan menganalisis data dan memanipulasi data

Mengubah tipe data variabel masing-masing atribut dari numpy.ndarray menjadi list

```python
Surat_list = [] # rubah dari numpy array ke list
for i in range(len(variabelSurat)):
  Surat_list.append(variabelSurat[i])

Ayat_list = [] # rubah dari numpy array ke list
for i in range(len(variabelAyat)):
  Ayat_list.append(variabelAyat[i])

Terjemahan_list = [] # rubah dari numpy array ke list
for i in range(len(variabelTerjemah)):
  Terjemahan_list.append(variabelTerjemah[i])
```

# Modelling

Metode yang digunakan untuk tahap pembuatan model machine learning untuk projek sistem temu kembali rekomendasi ayat Al-Qur'an relevan adalah cosine similarity.

## mamasukkan querry yang ingin dicari


```python
query = input("Masukkan querry dari judul paper yang ingin dicari : ")
```

## Melakukan Cosine similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# preprocessing dan vektorisasi documents
vectorizer = TfidfVectorizer()
vectorized_docs = vectorizer.fit_transform(Terjemahan_list)


# preprocessing dan vektorisasi querry
query_vector = vectorizer.transform([query])

# Menghitung cosine similarity antara kueri dengan masing-masing ayat Al-Qur'an
# Calculate cosine similarity between the query and each document
similarity_scores = cosine_similarity(query_vector, vectorized_docs)

# Menampilkan similarity scores untuk setiap dokumen

for doc_idx, score in enumerate(similarity_scores[0]):
    surat = Surat_list[doc_idx]
    ayat  = Ayat_list[doc_idx]
    terjemah = Terjemahan_list[doc_idx]

    print(f"Similarity dengan Al-Qur'an pada \nSurat ke-{surat}, \nAyat ke-{ayat}, \nyaitu sebesar: {score:.5f} \nIsi Terjemahan:\n{terjemah}\n\n")
```

## menyimpulkan kategori dari querry yang dimasukkan

```python
# dictionary variabel yang menyimpan list
dict = {'Urutan Surat' : Surat_list, 'Urutan Ayat': Ayat_list, 'Terjemahan Indonesia' : Terjemahan_list, 'nilai kesamaan': similarity_scores[0]}

df_Quran = pd.DataFrame(dict)
df_Quran = df_Quran[df_Quran['nilai kesamaan'] != 0]
Quran_sorted = df_Quran.sort_values(by=['nilai kesamaan'], ascending=False)

Quran_sorted[:]
```
